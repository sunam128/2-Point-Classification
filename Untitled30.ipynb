```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score #measure the accuracy of the algorithm
from sklearn.preprocessing import LabelEncoder
```

```python
#load the data
ds=pd.read_csv('Classif Data.csv')
ds.head()
```

```python
ds.info()
```

```python
#check for null values
ds.isnull().sum()
```

```python
#put the PROD column into labels/categories
ds['Prod']=pd.qcut(
    ds['Prod'],
    q=3,
    labels=["Low","Medium","High"] #or you can use [0,1,2,3] without using the 2nd method
)
#convert the labels to numerical/categorical data
le=LabelEncoder()
ds['Prod']=le.fit_transform(ds['Prod'])
```

```python
#feature extraction
y=ds['Prod']
x=ds.drop(columns='Prod',axis=1)
```

```python
#split the data into testing and training
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
#load the model
model=LogisticRegression(multi_class='multinomial',solver='lbfgs',max_iter=2000)
```

```python
#train the model
model.fit(x_train,y_train)
```

```python
#make predictions
predictions=model.predict(x_test)
```

```python
#compare predicted values with actual values
pd.DataFrame({'Actual values':y_test,'Predicted values':predictions})
```

```python
#evaluate the model's accuracy
accuracy_score(y_test,predictions)
```
