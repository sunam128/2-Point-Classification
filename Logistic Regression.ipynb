```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split #used to split data into testing and training
from sklearn.linear_model import LogisticRegression
```

```python
from google.colab import files
file=files.upload()
```

```python
#load and read the data
data=pd.read_csv('insurance_data.csv')
data.head()
```

```python
#getting the dataset information
data.info()
```

```python
#check for null values
data.isnull().sum()
```

```python
#checking or duplicated values
data.duplicated().sum()
```

```python
#feature extraction
y=data['bought_insurance']
x=data.drop('bought_insurance',axis=1)
```

```python
#split the data into testing and training
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
#load the algorithm
model=LogisticRegression()
```

```python
#train the model/algorithm
model.fit(x_train,y_train)
```

```python
#make predictions
predictions=model.predict(x_test)
```

```python
#compare the predicted values with actual values
pd.DataFrame({'Actual Values':y_test,"Predicted values":predictions})
```

```python
#check the accuracy of the algorithm/model
model.score(x_test,y_test)
```
